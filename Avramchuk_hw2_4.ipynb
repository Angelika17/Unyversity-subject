```python
def select_col(data_frame, column_names):
    new=data_frame.loc[:,column_names]
    return new

select_columns=["ім\'я","автор","ціна","рік" ]
tab_new=select_col(data, select_columns)
tab_2015=tab_new.loc[tab_new["рік"]==2015]

tab_2015[tab_2015["ціна"]==tab_2015["ціна"].max()]
```
